```python
data12["Ingresos (miles de pesos)"] = data12["Ingresos (miles de pesos)"].astype(str)
data12["Ingresos (miles de pesos)"] = data12["Ingresos (miles de pesos)"].str.replace(",","")
data12["Ingresos (miles de pesos)"] = data12["Ingresos (miles de pesos)"].str.replace(".","")
data12["Ingresos (miles de pesos)"] = pd.to_numeric(data12["Ingresos (miles de pesos)"])  

```

```python
data12.info()
```

Se analiza los ingresos por trimeste de cada año.

```python
plt.figure(figsize=(20,20))
sns.barplot(data = data12, x="Año",y="Ingresos (miles de pesos)", hue="Trimestre")
plt.title("Ingresos totales por año y trimestre")

plt.show()
```

Como se puede observar, los ingresos van en aumento constante. Esto es debido a que el costo de banda ancha es un poco más caro, pero ofrece mayores prestaciones.

```python
# se exporta a un archivo limpio para un futuro uso
data12.to_csv("Datasets/internet_ingresos_trimestre.csv")
```

```python
data13.info()
```

A comparación de otros datasets, se observa que los valores de algunas columnas son con valores decimales. Por lo tanto, es importante reemplazar la coma por un punto.


```python
for columna in data13:
    if columna == "Periodo":
        break
    else:
        data13[columna] = data13[columna].astype(str)
        data13[columna] = data13[columna].str.replace(",",".")
        data13[columna] = pd.to_numeric(data13[columna])




```

Se realiza un pequeño analisis sobre los accesos por cada 100 hogares.

```python
plt.figure(figsize=(20,20))
sns.barplot(data = data13, x="Año", y="Accesos por cada 100 hogares",errorbar=None,estimator="mean")
plt.title("Promedio de accesos por cada 100 hogares a través de los años", fontsize=30, color="blue")
plt.show()
```

Como se puede ver, por el crecimiento de las tecnologías, el promedio de los accesos van en aumento.

```python
# se exporta a un archivo limpio
data13.to_csv("Datasets/internet_penetracion_promedio_trimeste.csv")
```
